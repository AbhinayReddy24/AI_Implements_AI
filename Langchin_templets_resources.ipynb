```python
def read_markdown_file_raw(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            # Read the raw text content of the Markdown file
            markdown_content = file.read()

            return markdown_content

    except FileNotFoundError:
        print(f"Error: File not found - {file_path}")
        return None

# Replace 'your_file.md' with the actual path to your Markdown file
markdown_file_path = '/Users/abhinay/templets readme files/anthropic-iterative-search/README.md'

# Read and print the raw text content of the Markdown file
raw_content = read_markdown_file_raw(markdown_file_path)

if raw_content:
    print(raw_content)

```
